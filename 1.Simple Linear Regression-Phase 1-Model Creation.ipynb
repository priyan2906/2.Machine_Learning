```python
# Importing the pandas library to work with datasets (tables)
import pandas as pd
```

```python
# Reading the Salary Data CSV file into a dataframe named dataset
dataset = pd.read_csv("Salary_Data.csv")
```

```python
# Display the loaded dataset
dataset
```

```python
# Selecting the independent variable (input feature: Years of Experience)
independent = dataset[["Years of Experience"]]
# Display the independent variable (feature)
independent
```

```python
# Selecting the dependent variable (target/output: Salary)
dependent = dataset[["Salary"]]
# Display the dependent variable (target)
dependent
```

```python
'''Splitting data into training set and testing set
   test_size = 0.30 means 30% of data goes for testing
   random_state ensures same split every time (for reproducibility)'''
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# Importing Linear Regression model from sklearn
from sklearn.linear_model import LinearRegression
# Creating an object of LinearRegression model
regressor = LinearRegression()
# Training (fitting) the model using training data
regressor.fit(X_train,y_train)
```

```python
# Getting the slope (weight) value of the trained model
weight = regressor.coef_
# Display the learned weight(s)
weight
```

```python
# Getting the intercept (bias) value of the trained model
bias = regressor.intercept_
# Display the learned bias (intercept)
bias
```

```python
# Predicting salary using the test data
y_pred = regressor.predict(X_test)
```

```python
# Measuring accuracy using R² score
from sklearn.metrics import r2_score
# Finding model accuracy using R² score
r_score = r2_score(y_test,y_pred)
#Display r_score value 0 to 1, higher value will be good prediction
r_score
```

```python
# Saving the trained model into a file using pickle
import pickle
filename = "finalized_model_linear.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
# Loading the saved model back from file
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
```

```python
# Predicting salary for 15 years of experience
result = loaded_model.predict([[15]])
```
